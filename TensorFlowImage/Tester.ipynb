```python
import cv2
import tensorflow as tf
import numpy as np
from keras.models import load_model
loaded_model = load_model("testmodel.h5")
cap = cv2.VideoCapture(0)
img_height = 180
img_width = 180
class_names = ['kaan', 'muammer']

while True:
    _, frame = cap.read()
    frame2 = cv2.resize(frame, (img_width, img_height))
    img = tf.keras.preprocessing.image.img_to_array(frame2)
    img_array = tf.keras.utils.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0) # Create a batch
    predictions = loaded_model.predict(img_array)
    score = tf.nn.softmax(predictions[0])
    score_text = "{} - {:.2f}".format(class_names[np.argmax(score)], 100 * np.max(score))
    cv2.putText(frame, score_text, (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
    cv2.imshow("Frame", frame)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

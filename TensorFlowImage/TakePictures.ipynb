```python
import os
import pathlib
import shutil
data_dir = pathlib.Path(os.getcwd()+"\\images\\")
if os.path.exists(data_dir):
    shutil.rmtree(data_dir)
data_dir.mkdir()

#NOTE - yeni model için sıfırdan resim çekme işlemi yapılır
```

```python
import tkinter as tk
import cv2
from PIL import Image, ImageTk

cap = cv2.VideoCapture(0)
take_a_picture = False
picture_count = 0
def show_frame():
    global take_a_picture
    global picture_count
    total_picture = 150
    _, frame = cap.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGBA)
    if take_a_picture:
        folder_path = os.path.join(data_dir,entry.get())
        if not os.path.exists(folder_path):
            os.mkdir(folder_path)
        if picture_count<total_picture:
            _, frame = cap.read()
            file_path = os.path.join(folder_path, f"{picture_count}.jpg")
            cv2.imwrite(file_path, frame)
            picture_count+=1
        if picture_count>=total_picture:
            picture_count=0
            take_a_picture=False
    img = Image.fromarray(frame)
    imgtk = ImageTk.PhotoImage(image=img)
    lmain.imgtk = imgtk
    lmain.configure(image=imgtk)
    lmain.after(10, show_frame)

def get_images():
    global take_a_picture
    if not take_a_picture:
        take_a_picture=True

def train():
    print("train")

def quit():
    cap.release()
    top.destroy()

top = tk.Tk()
top.title("Welcome to Trainer")

lmain = tk.Label(top)
lmain.pack()
entry = tk.Entry(top, width=30, font=25)
entry.pack()

show_frame()

tk.Button(top, text="Take Pictures", command=lambda: get_images(), width=30, height=3, font=25).pack()
tk.Button(top, text="Train", command=lambda: train(), width=30, height=3, font=25).pack()
tk.Button(top, text="QUIT", command=lambda: quit(), width=30, height=3, font=25).pack()

top.mainloop()
```
